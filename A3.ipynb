```python
import tkinter as tk
import random
from soundplayer import *    # استفاده از کتابخانه ای که در فایل پایتونی ساختم
%reload_ext autoreload
%autoreload 2
import threading
import time
import queue



in_queue = []
counter = 100
thread_running = False

priority_q = queue.PriorityQueue()

# تابع برای اضافه کردن نوبت جدید :

def enqueue():

    global counter , thread_running
    customer_number = counter
    counter += 1  
    random_choice = random.randint(1, 5)  
    random_timing = random.randint(3, 6)  #سرویس دهی در حال حاضر بین 3 تا 6 ثانیه تنظیم شده
    priority_q.put((random_timing, customer_number, random_choice))  
    display()


    if not thread_running:
        thread_running = True
        t = threading.Thread(target=auto_dequeue,daemon=True)
        t.start()

# تابع برای حذف نوبت :

def dequeue():

    if priority_q.empty():
        label.config(text='صف خالی است')
    else:
        random_timing, customer, cabin = priority_q.get()
        text1 = f" شماره {customer} به باجه ی {cabin}"
        label.config(text=text1)
        
        play_sentence(customer, cabin)  

        display()
       

# تابع برای نمایش صف :

def display():
        
    queue_box.delete(0, tk.END) 
    for item in priority_q.queue:
        customer = item[1] 
        queue_box.insert(tk.END, f" شماره ی  {customer}") 



# تابع برای مدت زمان سرویس دهی : 

def auto_dequeue():

    global thread_running

    if priority_q.empty():
            thread_running = False
            return
    random_timing, customer, cabin = priority_q.get()

    root.after(0, lambda: label.config(text='لطفا منتظر بمانید'))
    root.after(random_timing * 1000, lambda: process_customer(customer, cabin))


def process_customer(customer, cabin):

    text1 = f" شماره {customer} به باجه ی {cabin}"
    label.config(text=text1)

    play_sentence(customer, cabin)
    display()

    if priority_q.empty():
        global thread_running
        thread_running = False  
    else:
        auto_dequeue() 



root = tk.Tk()
root.title("سیستم نوبت دهی بانکی")

frame = tk.Frame(root)
frame.pack(pady=20)

button1 = tk.Button(frame, text="نوبت جدید", command=enqueue)
button1.grid(row=0, column=0, padx=10)
#button2 = tk.Button(frame, text="بعدی", command=dequeue)
#button2.grid(row=0, column=1, padx=10)

label = tk.Label(root, text="برای گرفتن نوبت، گزینه ی بالا را انتخاب نمایید")
label.pack(pady=20)

queue_box = tk.Listbox(root, height=10, width=40)
queue_box.pack()

root.geometry("400x400")
root.mainloop()
```
